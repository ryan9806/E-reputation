```python
from pyspark.ml.feature import Tokenizer, RegexTokenizer
from pyspark.sql.functions import col, udf
from pyspark.sql.types import IntegerType
```

```python
tokenizer = Tokenizer(inputCol="text", outputCol="words")
```

```python
regexTokenizer = RegexTokenizer(inputCol="text", outputCol="words", pattern="\\W")
```

```python
countTokens = udf(lambda words: len(words), IntegerType())
```

```python
tokenized = tokenizer.transform(df1)
```

```python
tokenized.select("text", "words")\
    .withColumn("tokens", countTokens(col("words"))).show(1,truncate=False)
```

```python
regexTokenized = regexTokenizer.transform(df1)
```

```python
regexTokenized.select("text", "words") \
    .withColumn("tokens", countTokens(col("words"))).show(truncate=False)
```

```python
from pyspark.ml.feature import NGram
```

```python
ngram = NGram(n=2, inputCol="words", outputCol="ngrams")
```

```python
ngramDataFrame = ngram.transform(tokenized)
ngramDataFrame.select("ngrams").show(truncate=False)
```

```python
ngramDataFrame.show(1)
```

```python
from pyspark.ml.feature import HashingTF, IDF, Tokenizer
from pyspark.ml import Pipeline
```

```python
tokenizer = Tokenizer(inputCol="text", outputCol="testee")
wordsData = tokenizer.transform(ngramDataFrame)
```

```python
wordsData
```

```python
wordsData.select("testee").show(truncate=False)
```

```python
hashingTF = HashingTF(numFeatures=262144,inputCol="testee", outputCol="rawFeatures")
featurizedData = hashingTF.transform(wordsData)
```

```python
#rescaledData.groupby('label').count().collect()
```

```python
idf = IDF( inputCol="rawFeatures", outputCol="features")
idfModel = idf.fit(featurizedData)
rescaledData = idfModel.transform(featurizedData)

rescaledData.select("label", "features").show()
```

```python
#rescaledData.select("features").show(truncate=False)
```

```python
rescaledData.groupby('label').count().collect()
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
from pyspark.sql import Row
from pyspark.ml.classification import LogisticRegression
```

```python
from pyspark.ml.linalg import Vectors
```

```python
lr = LogisticRegression(featuresCol='features',labelCol='label')
```

```python
# Fit the model
lrModel = lr.fit(rescaledData)
```

```python
pred = lrModel.transform(rescaledData)
```

```python
pred.select("label","prediction").show()
```

```python
lr.getRegParam()
```
